```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import math

from scipy import stats
from sklearn.preprocessing import StandardScaler
from factor_analyzer import FactorAnalyzer
```

## Demographics - answered questionnaires by year, age and gender

```python
fig, ax = plt.subplots(1, 3, figsize = (18,5))
fig.suptitle('Responses by...', size=16, y = 1.02)


plots = ('quest_year', 'age', 'gender')
c = 0
for column in plots:
    column_count = df[column].value_counts().to_dict()
    ax[c].bar(column_count.keys(), column_count.values())
    ax[c].set_title(column, size=16, y = 1.02)
    c = c+1
```

##  Healthy BMI in percent of respondents by year, age and gender

```python
fig, ax = plt.subplots(1, 3, figsize = (18,5))
fig.suptitle('Healthy BMI in percent by... (red line: average)', size=16, y = 1.02)


plots = ('quest_year', 'age', 'gender')
c = 0
for column in plots:
    avg = df.groupby(column)['healthy_bmi'].mean().to_dict()
    ax[c].bar(avg.keys(), avg.values())
    ax[c].axhline(df['healthy_bmi'].mean(), color='red', linewidth=2)
    ax[c].set_title(column, size=16, y = 1.02)
    ax[c].set_ylim(0,1)
    c = c+1
```

## Mean values of all features for the two groups of healthy_bmi

```python
fig, ax = plt.subplots(4, 5, figsize = (20,20))
fig.suptitle('Average values of ... by healthy_bmi', size=16)


plots = ('fruits_veggies', 'daily_stress', 'places_visited', 'core_circle',
       'supporting_others', 'social_network', 'achievement', 'donation',
       'todo_completed', 'flow', 'daily_steps', 'live_vision', 'sleep_hours',
       'lost_vacation', 'daily_shouting', 'sufficient_income',
       'personal_awards', 'time_for_passion', 'weekly_meditation')
a = 0
b = 0
for column in plots:
    avg = df.groupby('healthy_bmi')[column].mean().to_dict()
    ax[a,b].bar(avg.keys(), avg.values())
    ax[a,b].axhline(df[column].mean(), color='red', linewidth=2)
    ax[a,b].set_title(column, size=16, y = 1.02)
    ax[a,b].set_ylim(df[column].mean()-1,df[column].mean()+1)
    if b < 4:
        b = b+1
        a = a
    else:
        b = 0
        a = a+1
```

### Checking the thought as "good" variables, that have a lower mean for healthy bmi respondents than for non healthy bmi respondents

```python
fig, ax = plt.subplots(5, 2, figsize = (20,20))

plots = ('supporting_others', 'social_network', 'donation', 'flow',
       'personal_awards')
a = 0
for column in plots:

    ax[a,0].hist(df[df['healthy_bmi'] == 0][column], bins = 11, density = True)
    ax[a,0].set_title('not an healthy bmi '+column, size=16, y = 1.02)
    
    ax[a,1].hist(df[df['healthy_bmi'] == 1][column], bins = 11, density = True)
    ax[a,1].set_title('healthy bmi '+column, size=16, y = 1.02)
    
    a = a+1
```

### distributions are very similar in both groups but the people with a bmi > 25 (not healthy) have more maximum ratings for the questions.

## Finding dimensions behind features with factor analysis

```python
#selecting numeric features, excluding sufficient income, which is binary, work-life-balance score, which is calculated from the features,
#healthy bmi, also binary and year and month of questionnaire)
features = df.select_dtypes('number').drop(['sufficient_income', 'work_life_balance_score', 'healthy_bmi', 'quest_year',
       'quest_month'], axis = 1)
```

```python
features
```

```python
#Standardize variables for factor analysis
transformer = StandardScaler().fit(features)
features = pd.DataFrame(transformer.transform(features), columns=features.columns)
```

```python
#Bartlett test - are variables uncorrelated?
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value,p_value=calculate_bartlett_sphericity(features)
print('Chi_square:', chi_square_value, '; p_value:', p_value)
```

```python
#KMO and MSA test - measure of sampling adequacy for whole model (KMO, last number) and 
# for each feature (MSA, list of 18 numbers) - should be at least 0.5, best between 0.85 and 1 

from factor_analyzer.factor_analyzer import calculate_kmo
kmo_model=calculate_kmo(features)
print(kmo_model)
```

### KMO and all MSA are above 0.5, there are 4 features below 0.8 --> creating a second feature set,keeping only features with MSA < 0.8

```python
features_red = features.drop(['daily_stress', 'sleep_hours', 'lost_vacation', 'daily_shouting'], axis = 1)

#KMO and MSA test - measure of sampling adequacy for whole model (KMO, last number) and 
# for each feature (MSA, list of 18 numbers) - should be at least 0.5, best between 0.85 and 1 
kmo_model=calculate_kmo(features_red)
print(kmo_model)
```

## All features

### Choosing number of factors

```python
#running factor analysis with as many factors as features, to check eigenvalues and cumulative variance
fa = FactorAnalyzer(n_factors=18, rotation=None, method='principal')
fa.fit(features)
```

### Elbow method to choose number of factors

```python
# saving eigenvalues in a list
eigenvalues = list(fa.get_eigenvalues()[0])
```

```python
#plotting eigenvalues and number of factors
plt.plot(range(1,19), eigenvalues, color = 'blue', linestyle = 'dashed', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title('Eigenvalues vs. Number Components')
plt.gca().invert_yaxis()
plt.xlabel('#Components')
plt.ylabel('Eigenvalues')
plt.show()
```

#### Elbow method would suggest either 2 or 4 features

### Kaiser criterion and cumulative variance: Kaiser criterion: choose number of factors with eigenvalue > 1, Variance: cumulative explained variance should be at least 80%

```python
variance_expl = np.round(fa.get_factor_variance(), decimals=2)
print('Eigenvalue/Proportional/Cumulative Variance:')
print(variance_expl)
```

#### Kaiser criterion would suggest 4 factors. However,  to reach 80% cumulative explained variance, 12 factors would be needed. 4 factors only explain 44% of the variance, the 2 factor solution would explain only 31%. 

### 4 Factor solution

```python
fa4 = FactorAnalyzer(n_factors=4, method='principal', rotation='varimax')
fa4.fit(features)

#get the loadings of variables on factors
print(pd.DataFrame(fa4.loadings_,index=features.columns,
                   columns = ['Factor 1', 'Factor 2', 'Factor 3', 'Factor 4']))
```

#Factor loadings sorted


#### There are two variables with cross-loadings on more than 1 factor (with less than 0.2 difference): todo_completed (Factor 3 and 4) and places visited (Factor 1 and 4).

```python
#trying default setting (minres and oblique rotation)
fa4_mr = FactorAnalyzer(n_factors=4, method='minres', rotation='promax')
fa4_mr.fit(features)

#get the loadings of variables on factors
print(pd.DataFrame(fa4_mr.loadings_,index=features.columns,
                   columns = ['Factor 1', 'Factor 2', 'Factor 3', 'Factor 4']))
```

#### Factor loadings sorted


#### --> factors are basically the same (same features) but the loadings are less distinct/lower than with pca and orthogonal rotation

### 2 Factor solution

```python
fa2 = FactorAnalyzer(n_factors=2, method='principal', rotation='varimax')
fa2.fit(features)

#get the loadings of variables on factors
print(pd.DataFrame(fa2.loadings_,index=features.columns,
                   columns = ['Factor 1', 'Factor 2']))
```

#### Factor loadings sorted


```python
# Two factor solution for minres and oblique rotation
fa2_mr = FactorAnalyzer(n_factors=2, method='minres', rotation='promax')
fa2_mr.fit(features)

#get the loadings of variables on factors
print(pd.DataFrame(fa2_mr.loadings_,index=features.columns,
                   columns = ['Factor 1', 'Factor 2']))
```

#### Factor loadings sorted



#### again, same structur but factor loadings are higher with PCA

### 3 Factor solution - as 2 Factor solution is hard to interpret, also checking 3 Factor solution

```python
fa3 = FactorAnalyzer(n_factors=3, method='principal', rotation='varimax')
fa3.fit(features)

#get the loadings of variables on factors
print(pd.DataFrame(fa3.loadings_,index=features.columns,
                   columns = ['Factor 1', 'Factor 2', 'Factor 3']))
```

#### Factor loadings sorted


### If taking all features into factor analysis, the 4 factor solution is the best solution. Factors can be interpreted as: 1: social commitment & recognition, 2: stress level, 3: self-fulfilment and 4: health & wellbeing

## Factor analysis with reduced features

### Choosing number of factors

```python
#running factor analysis with as many factors as features, to check eigenvalues and cumulative variance
fa = FactorAnalyzer(n_factors=14, rotation=None, method='principal')
fa.fit(features_red)
```

### Elbow method to choose number of factors

```python
# saving eigenvalues in a list
eigenvalues = list(fa.get_eigenvalues()[0])
```

```python
#plotting eigenvalues and number of factors
plt.plot(range(1,15), eigenvalues, color = 'blue', linestyle = 'dashed', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title('Eigenvalues vs. Number Components')
plt.gca().invert_yaxis()
plt.xlabel('#Components')
plt.ylabel('Eigenvalues')
plt.show()
```

#### Elbow method would suggest  2 features

### Kaiser criterion and cumulative variance

```python
variance_expl = np.round(fa.get_factor_variance(), decimals=2)
print('Eigenvalue/Proportional/Cumulative Variance:')
print(variance_expl)
```

#### Kaiser criterion would suggest 3 factors. However,  to reach 80% cumulative explained variance, 10 factors would be needed. 3 factors only explain 43% of the variance, the 2 factor solution would explain only 35%. 

### 3 Factor solution

```python
fa_red3 = FactorAnalyzer(n_factors=3, method='principal', rotation='varimax')
fa_red3.fit(features_red)

#get the loadings of variables on factors
print(pd.DataFrame(fa_red3.loadings_,index=features_red.columns,
                   columns = ['Factor 1', 'Factor 2', 'Factor 3']))
```

#### Factor loadings sorted


#### The 3 Factor solution with reduced features corresponds to the 4 Factor solution with all features minus the stressor factor. Again, todo_completed has crossloading on two factors (Factor 1 and 2)

### 2 Factor solution

```python
fa_red2 = FactorAnalyzer(n_factors=2, method='principal', rotation='varimax')
fa_red2.fit(features_red)

#get the loadings of variables on factors
print(pd.DataFrame(fa_red2.loadings_,index=features_red.columns,
                   columns = ['Factor 1', 'Factor 2']))
```

#### Factor loadings sorted


#### 2 Factor solution with reduced features can be interpreted as Facor 1: self-fulfilment & social commitment, Factor 2: health & well-being

## 4 Factor solution with all features and 3 Factor solution with reduced features are best solutions. Storing factor values in dataframe

```python
# savingfactor values for each row in dataframe
factors4=pd.DataFrame(fa4.transform(features), columns=['F1_social', 'F2_stress', 'F3_fulfilment', 'F4_health'])
factors3=pd.DataFrame(fa_red3.transform(features_red), columns=['F_red1_fulfilment', 'F_red2_health', 'F_red3_social'])
df=pd.concat([df, factors4, factors3], axis=1)
df.info()
```

```python
factorscores = df[['F1_social', 'F2_stress', 'F3_fulfilment', 'F4_health',
       'F_red1_fulfilment', 'F_red2_health', 'F_red3_social']]
```
