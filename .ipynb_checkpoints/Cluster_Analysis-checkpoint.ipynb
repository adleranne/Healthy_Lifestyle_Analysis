```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import hdbscan
from sklearn.metrics import silhouette_score

from collections import defaultdict
```

```python
#import dataframe including factors from EDA

df = pd.read_csv('lifestyle_inclfactors.csv', index_col=0)
```

```python
#check dataframe

df.head(5)
```

```python
df.columns
```

# Clustering by Factors

```python
#defining three different set of features - 4 factor and 3 factor solutions from PCA, using the minmax scaled scores
# and all single features
X_4 = df[['sc_social', 'sc_stress', 'sc_fulfilment', 'sc_health']]
X_3 = df[['sc2_fulfilment','sc2_health', 'sc2_social']]
X_all = df[['fruits_veggies', 'daily_stress', 'places_visited', 'core_circle',
       'supporting_others', 'social_network', 'achievement', 'donation',
       'todo_completed', 'flow', 'daily_steps', 'live_vision', 'sleep_hours',
       'lost_vacation', 'daily_shouting', 'personal_awards', 'time_for_passion', 'weekly_meditation']]

# single features need to be transformed
transformer = MinMaxScaler().fit(X_all)
X_all = transformer.transform(X_all)
```

# KMeans

```python
# Choosing number of clusters with KMeans

feature_sets = [X_4, X_3, X_all]
K = range(2, 11)
a = 0

fig, ax = plt.subplots(len(feature_sets), 2, figsize = (18,len(feature_sets)*5))

for X in feature_sets:
    
    inertia = []
    silhouette = []
    
    for k in K:
        kmeans = KMeans(n_clusters=k, n_init = 30, max_iter = 1000, algorithm = 'full',
                    random_state=1234)
        kmeans.fit(X)
        inertia.append(kmeans.inertia_)
        silhouette.append(silhouette_score(X, kmeans.predict(X)))

    ax[a,0].plot(K, inertia, 'bx-')
    ax[a,0].set_xticks(np.arange(min(K), max(K)+1, 1.0))
    ax[a,0].set_title('Inertia')
    
    ax[a,1].plot(K, silhouette, 'bx-')
    ax[a,1].set_xticks(np.arange(min(K), max(K)+1, 1.0))
    ax[a,1].set_title('silhouette score')
    
    a = a+1
```

### For both factor feature sets 4 clusters are the optimum - the three factor solution has better values concerning inertia and silhoutte score. All single features as input variables don't really show an elbow for inertia and highest silhoutte score is for 2 cluster solution (but lower score than for X_4 and X_3).

```python
# defining a function to view average values of certain features by cluster

def view_clustervalues(cluster, features):
    
    # get the mean values for each feature in the different clusters and store in one concatenated dicionairy
    cluster_values = defaultdict(list)
    for feature in features:
        values = df.groupby(cluster)[feature].mean().to_dict()
        for key, value in values.items():
            cluster_values[key].append(value)
    
    # plot mean values per cluster
    fig, ax = plt.subplots(1, df[cluster].nunique(), figsize = (18,5))
    c = 0
    for clust in cluster_values.keys():
        bars = features
        height = (cluster_values[clust])
        ax[c].bar(bars, height)
        ax[c].set_ylim(-1.5,1.5)
        c = c+1
```

## 4 Clusters with X_4

```python
# Clustering 4 clusters with X_4:
kmeans = KMeans(n_clusters=4, n_init = 30, max_iter = 1000, algorithm = 'full', random_state=1234)
kmeans.fit(X_4)

# Predicting / assigning the clusters:
clusters = kmeans.predict(X_4)

# Check the size of the clusters
print(pd.Series(clusters).value_counts().sort_index())

# Storing clusters in dataframe
df["cluster"] = clusters
```

```python
# viewing average value of 4 original factors for each cluster
features = ['F1_social', 'F2_stress', 'F3_fulfilment', 'F4_health']
view_clustervalues('cluster', features)
```

## 4 Clusters with X_3

```python
# Clustering 4 clusters with X_3:
kmeans = KMeans(n_clusters=4, n_init = 30, max_iter = 1000, algorithm = 'full', random_state=1234)
kmeans.fit(X_3)

# Predicting / assigning the clusters:
clusters = kmeans.predict(X_3)

# Check the size of the clusters
print(pd.Series(clusters).value_counts().sort_index())

# Storing clusters in dataframe
df["cluster"] = clusters
```

```python
# viewing average value of 3 original factors for each cluster
features = ['F_red1_fulfilment', 'F_red2_health', 'F_red3_social']
view_clustervalues('cluster', features)
```

# Agglomerative Clustering

```python
# Choosing number of clusters

feature_sets = [X_4, X_3, X_all]
K = range(2, 11)
a = 0

fig, ax = plt.subplots(len(feature_sets), 1, figsize = (18,len(feature_sets)*5))

for X in feature_sets:
    silhouette = []
    
    for k in K:
        model = AgglomerativeClustering(n_clusters=k)
        model.fit(X)
        silhouette.append(silhouette_score(X, model.fit_predict(X)))
    
    ax[a].plot(K, silhouette, 'bx-')
    ax[a].set_xticks(np.arange(min(K), max(K)+1, 1.0))
    ax[a].set_title('silhouette score')
    
    a = a+1
```

### Agglomerative Clustering has the highest silhoutte score for X_3 feature set with 4 clusters. Even though, silhoutte score is lower than for KMeans clustering

```python
# Clustering 4 clusters with X_3:
aggcluster_X3 = AgglomerativeClustering(n_clusters = 4).fit(X_3)

# Predicting / assigning the clusters:
clusters = aggcluster_X3.fit_predict(X_3)

# Check the size of the clusters
print(pd.Series(clusters).value_counts().sort_index())

# Storing clusters in dataframe
df["cluster"] = clusters
```

```python
# viewing average value of 3 original factors for each cluster
features = ['F_red1_fulfilment', 'F_red2_health', 'F_red3_social']
view_clustervalues('cluster', features)
```

```python
# viewing average value of 3 factors minmax scaled for each cluster
features = ['sc2_fulfilment', 'sc2_health', 'sc2_social']
view_clustervalues('cluster', features)
```

# HDBSCAN clustering

```python
# clustering with X_4
model = hdbscan.HDBSCAN().fit(X_4)

# getting number of clusters
print('clusters: ', model.labels_.max()+1)

# getting number of cases which are noise (not in a cluster)
print('not clustered cases: ', pd.Series(model.labels_).value_counts()[-1])
print('not clustered cases in percent: ', (pd.Series(model.labels_).value_counts()[-1])/len(df))
```

```python
# clustering with X_3
model = hdbscan.HDBSCAN().fit(X_3)

# getting number of clusters
print('clusters: ', model.labels_.max()+1)

# getting number of cases which are noise (not in a cluster)
print('not clustered cases: ', pd.Series(model.labels_).value_counts()[-1])
print('not clustered cases in percent: ', (pd.Series(model.labels_).value_counts()[-1])/len(df))
```

```python
# clustering with X_all
model = hdbscan.HDBSCAN().fit(X_all)

# getting number of clusters
print('clusters: ', model.labels_.max()+1)

# getting number of cases which are noise (not in a cluster)
print('not clustered cases: ', pd.Series(model.labels_).value_counts()[-1])
print('not clustered cases in percent: ', (pd.Series(model.labels_).value_counts()[-1])/len(df))
```

### HDBSCAN is not able to identify clusters, most datapoints (between 71% and 95%) are considered 'noise' and don't belong to a cluster
