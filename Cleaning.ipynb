```python
import pandas as pd
import numpy as np
```

```python
# import dataset
df = pd.read_csv('Wellbeing_and_lifestyle_data_Kaggle.csv')
```

```python
df.head()
```

## Cleaning

```python
#rename column headers, all lower case
df.columns = [i.lower() for i in df.columns]
```

```python
#check number of rows, missings and column types
df.info()
```

```python
#daily stress should be numeric, check values
df['daily_stress'].value_counts()
```

```python
#check datarow with '1/1/00' score
df[df['daily_stress'] == '1/1/00']
```

```python
#since rest of row is okay, replacing 1/1/00 with most common value of the both values 0 and 1 - which is 1:
df.loc[df['daily_stress'] == '1/1/00', 'daily_stress'] = '1'
```

```python
#change datatype of column to integer
df['daily_stress'] = df['daily_stress'].astype(int)
df['daily_stress'].value_counts()
```

```python
#check values of integer columns
pd.set_option('display.max_columns', None)
df.describe()
```

```python
#function to recode the two binary variables (bmi_range and sufficient_income) from 1/2 into 0/1:

def recode_binary(i):
    if i == 2:
        return 0
    else:
        return i
```

```python
#bmi_range: 1 = below 25, 2 = above 25, naming recoded variable 'healthy_bmi', with 0 = no (above 25) and 1 = yes (below 25):
df['healthy_bmi'] = df['bmi_range'].apply(recode_binary)
```

```python
#check result
pd.crosstab(df['bmi_range'], df['healthy_bmi'])
```

```python
#sufficient_income: 1 = not hardly sufficient, 2 = sufficient, '2' should be coded as 'yes'(1) and 1 as 'no'(0)
#new function needed
def recode_binary2(i):
    if i == 1:
        return 0
    else:
        return 1
```

```python
#apply on sufficient_income and check value_counts before/after
print(df['sufficient_income'].value_counts())
df['sufficient_income'] = df['sufficient_income'].apply(recode_binary2)
print(df['sufficient_income'].value_counts())
```

```python
#checking two object columns age and gender:
df['age'].value_counts()
```

```python
## assuming 'less than 20' is including 20 year old, since other ranges include upper boundary of range.
#change label to '20 or younger'
df['age'] = df['age'].apply(lambda x: x.replace('Less than 20', '20 or younger'))
```

```python
df['gender'].value_counts()
```

```python
#check timestamp
df['timestamp'].value_counts()
```

```python
#converting into datetime format and extracting year and month as seperate columns
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['quest_year'] = df['timestamp'].dt.year
df['quest_month'] = df['timestamp'].dt.month
```

```python
#check year and month column
print(df['quest_year'].value_counts())
print(df['quest_month'].value_counts())
```

```python
#drop columns timestamp and bmi_range
df.drop(['timestamp', 'bmi_range'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.columns = ['fruits_veggies', 'daily_stress', 'places_visited', 'core_circle',
       'supporting_others', 'social_network', 'achievement', 'donation',
       'todo_completed', 'flow', 'daily_steps', 'live_vision', 'sleep_hours',
       'lost_vacation', 'daily_shouting', 'sufficient_income',
       'personal_awards', 'time_for_passion', 'weekly_meditation', 'age',
       'gender', 'work_life_balance_score', 'healthy_bmi', 'quest_year',
       'quest_month']
```

```python
#save cleaned datafile
df.to_csv('lifestyle_cleaned.csv')
```
